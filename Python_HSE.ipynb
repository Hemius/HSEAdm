```python
# Проверка счастливого билета

# Пример значения билета
number = 123321  # Можно изменить на любое шестизначное число

# Преобразуем число в строку и проверим, что оно шестизначное
num_str = str(number)

# Разделим на две части
first_half = num_str[:3]
second_half = num_str[3:]

# Посчитаем суммы
sum_first = sum(int(digit) for digit in first_half)
sum_second = sum(int(digit) for digit in second_half)

# Проверим условие
if sum_first == sum_second:
    print("Счастливый билет")
else:
    print("Несчастливый билет")
```

```python
# Проверка високосного года

# Ввод года
year = int(input("Введите номер года: "))

# Проверка условия и вывод
if year == 2020:
    print("Високосный год")
elif year == 2019:
    print("Обычный год")
else:
    print("Год не предусмотрен заданием")
```
